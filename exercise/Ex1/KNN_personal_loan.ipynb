# ***K - Nearest Neightbor Classifier***

# Bài toán phân loại sử dụng KNN 

Mục tiêu: 

- Xây dựng được mô hình KNN sử dụng thư viện sklearn. 
- Ứng dụng, hiểu cách áp dụng mô hình KNN vào giải quyết bài toán thực tế (vd: phân loại) 
- Sử dụng độ đo Accuracy, Precision, Recall để làm độ đo đánh giá chất lượng mô hình. 

Vấn đề: 
- Có một tập các dữ liệu không có nhãn, làm sao để biết dữ liệu này là thuộc về nhãn nào.
- => Xây dựng mô hình học máy có thể phân loại.  

Dữ liệu: 
- Dữ liệu Bank Personal Loan Modelling
- Xem thêm: https://www.kaggle.com/teertha/personal-loan-modeling

Bài toán: 
- Input: 1 mẫu dữ liệu $X = [x_1, x_2, ... x_n]$
- Output: nhãn $y$ là 0 hoặc 1 

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

%matplotlib inline
```

# **1.** Mô hình KNN
Sử dụng thư viện sklearn để xây dựng mô hình
- `KNeighborsClassifier(n_neighbors = 10, metric = 'minkowski', p = 2)`
- Số láng giềng: n_neighbors = 5
- Độ đo khoảng cách: Euclide p = 2

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 10, metric = 'minkowski', p = 2, weights = 'distance')
knn_classifier.fit(X_train, y_train)
```

# **2.** Testing KNN model

## Đánh giá theo các độ đo

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score

print("Testing...\n")
y_pred_knn = knn_classifier.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred_knn))
print('Precision: ', precision_score(y_test, y_pred_knn))
print('Recall: ', recall_score(y_test, y_pred_knn))
```

# **3.** Lựa chọn mô hình

## Lựa chọn số lượng láng giềng
- Thay đổi số lượng láng giềng tìm giá trị cho kết quả phân loại tốt nhất

```python
#train
n_neighbors = np.arange(1, 21)
test_accuracy = np.empty(len(n_neighbors))
test_precision = np.empty(len(n_neighbors))
test_recall = np.empty(len(n_neighbors))

for i in range(1,21):
    knn_model = KNeighborsClassifier(n_neighbors = i, metric = 'minkowski', p = 2, weights = 'distance')
    knn_model.fit(X_train, y_train)
    y_pred_knn = knn_model.predict(X_test)
    test_accuracy[i-1] = accuracy_score(y_test, y_pred_knn)
    test_precision[i-1] = precision_score(y_test, y_pred_knn)
    test_recall[i-1] = recall_score(y_test, y_pred_knn)
```

```python
data_wide = {'Neighbors': n_neighbors,
             'Accuracy': test_accuracy,
             'Precision': test_precision,
             'Recall': test_recall}
data_wide = pd.DataFrame(data_wide)
data_wide = pd.melt(data_wide, id_vars = 'Neighbors', var_name = 'Metric', value_name = 'Validation')
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data = data_wide, x = 'Neighbors', y = 'Validation', hue = 'Metric', marker = 'o')

plt.title('KNN: Varying Number of Neighbors')
plt.xlabel('Number of Neighbors')
plt.ylabel('Validation Metric')
plt.xticks(np.arange(1, 21, 1))
plt.legend(loc='best')
plt.grid()
plt.show()
```

## Lựa chọn thuộc tính
- Các thuộc tính: 'ID', 'Age', 'Experience', 'Income', 'ZIP Code', 'Family', 'CCAvg', 'Education', 'Mortgage', 'Personal Loan', 'Securities Account', 'CD Account', 'Online', 'CreditCard'.
- Thử loại bỏ từng thuộc tính ra khỏi dữ liệu xem chúng ảnh hưởng như thế nào tới kết quả phân loại.
- Các thuộc tính nào nên được sử dụng để cho kết quả phân loại tốt nhất ?

```python
X_train.columns
```

```python
k = 2
features = ['None', 'Age', 'Experience', 'Income', 'ZIP Code', 'Family', 'CCAvg', 'Education', 'Mortgage', 'Securities Account', 'CD Account', 'Online', 'CreditCard']
accuracy = []
precision = []
recall = []
for feature in features:
    X_train_copy = X_train.copy()
    X_test_copy = X_test.copy()

    if feature != 'None':
        X_train_copy = X_train_copy.drop(feature, axis=1)
        X_test_copy = X_test_copy.drop(feature, axis=1)

    knn_model = KNeighborsClassifier(n_neighbors = k, metric = 'minkowski', p = 2, weights = 'distance')

    knn_model.fit(X_train_copy, y_train)
    y_pred_knn = knn_model.predict(X_test_copy)

    accuracy.append(accuracy_score(y_test, y_pred_knn))
    precision.append(precision_score(y_test, y_pred_knn))
    recall.append(recall_score(y_test, y_pred_knn))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))

axes[0].plot(features, accuracy, marker='o')
axes[0].set_title('Accuracy vs Features Removed')
axes[0].set_xlabel('Features Removed')
axes[0].set_ylabel('Accuracy')
axes[0].set_xticklabels(features, rotation=45)

axes[1].plot(features, precision, marker='o')
axes[1].set_title('Precision vs Features Removed')
axes[1].set_xlabel('Features Removed')
axes[1].set_ylabel('Precision')
axes[1].set_xticklabels(features, rotation=45)
                        
axes[2].plot(features, recall, marker='o')
axes[2].set_title('Recall vs Features Removed')
axes[2].set_xlabel('Features Removed')
axes[2].set_ylabel('Recall')
axes[2].set_xticklabels(features, rotation=45)

plt.tight_layout()
plt.show()
```

## Lựa chọn hàm tính khoảng cách
- Hàm tính khoảng cách: minkowski, manhattan, euclidean, chebyshev
- Hàm tính khoảng cách nào là tốt nhất cho bài toán này ?

```python
X_test = X_test.drop('Online', axis=1)
X_train = X_train.drop('Online', axis=1)

metrics = ['minkowski', 'manhattan', 'euclidean', 'chebyshev']
acc = []
pre = []
rec = []
for metric in metrics:
    knn_model = KNeighborsClassifier(n_neighbors = 10, metric = metric, weights = 'distance')
    knn_model.fit(X_train, y_train)
    y_pred_knn = knn_model.predict(X_test)
    acc.append(accuracy_score(y_test, y_pred_knn))
    pre.append(precision_score(y_test, y_pred_knn))
    rec.append(recall_score(y_test, y_pred_knn))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))

axes[0].bar(metrics, acc)
axes[0].set_title('Accuracy vs Distance Metric')
axes[0].set_xlabel('Distance Metric')
axes[0].set_ylabel('Accuracy')
axes[0].axhline(y=max(acc), color='red', linestyle='--', linewidth=2)

axes[1].bar(metrics, pre)
axes[1].set_title('Precision vs Distance Metric')
axes[1].set_xlabel('Distance Metric')
axes[1].set_ylabel('Precision')
axes[1].axhline(y=max(pre), color='red', linestyle='--', linewidth=2)


axes[2].bar(metrics, rec)
axes[2].set_title('Recall vs Distance Metric')
axes[2].set_xlabel('Distance Metric')
axes[2].set_ylabel('Recall')
axes[2].axhline(y=max(rec), color='red', linestyle='--', linewidth=2)

plt.tight_layout()
plt.show()
```
